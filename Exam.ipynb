а1 = (1, 8, -1)

Исследуем систему векторов:

$$ \vec{a}_1 = \begin{pmatrix} 1; \ 8; \ -1 \end{pmatrix} $$

$$ \vec{a}_2 = \begin{pmatrix} -2; \ 3; \ 3 \end{pmatrix} $$

$$ \vec{a}_3 = \begin{pmatrix} 4; \ -11; \ 9 \end{pmatrix} $$

Матричное представление системы векторов:
формируем матрицу \( A \) из данных векторов:
   $
   A = \begin{pmatrix}
   1 & -2 & 4 \\
   8 & 3 & -11 \\
   -1 & 3 & 9
   \end{pmatrix}
   $

Решение:
Для определения линейной зависимости/независимости векторов составим определитель матрицы из этих векторов:



Вычислим определитель по правилу Саррюса:
$$ \det A = 1 \cdot 3 \cdot 9 + (-2) \cdot (-11) \cdot (-1) + 4 \cdot 8 \cdot 3 - 4 \cdot 3 \cdot (-1) - 1 \cdot (-11) \cdot 3 - (-2) \cdot 8 \cdot 9 $$

Выполним умножение:
$1 \cdot 3 \cdot 9 = 27$
$(-2) \cdot (-11) \cdot (-1) = -22$
$4 \cdot 8 \cdot 3 = 96$
$-4 \cdot 3 \cdot (-1) = 12$
$-1 \cdot (-11) \cdot 3 = 33$
$-(-2) \cdot 8 \cdot 9 = 144$
Сложим все члены: $$ \det A = 27 - 22 + 96 + 12 + 33 + 144 = 290 $$
Вывод:
Так как определитель матрицы не равен нулю ($\det A = 290 \neq 0$), то данные векторы являются линейно независимыми. Это означает, что ни один из векторов нельзя представить как линейную комбинацию других векторов системы.

При каких значениях параметра a система является линейно зависимой?

a1 = (1; 0; 2; -3),  
a2 = (2; 2; -4; a),  
a3 = (3; 1; 2; -5)

Для того чтобы система векторов $\mathbf{a}_1 = (1, 0, 2, -3)$, $\mathbf{a}_2 = (2, 2, -4, a)$, $\mathbf{a}_3 = (3, 1, 2, -5)$ была линейно зависимой, должно существовать нетривиальное решение уравнения:

$
c_1 \mathbf{a}_1 + c_2 \mathbf{a}_2 + c_3 \mathbf{a}_3 = \mathbf{0}
$

где $\mathbf{0}$ — нулевой вектор, и хотя бы один из коэффициентов $c_1, c_2, c_3$ отличен от нуля

Это эквивалентно тому, что матрица, составленная из этих векторов, должна иметь ранг, меньший числа векторов. 
Составим матрицу:

$
A = \begin{bmatrix}
1 & 2 & 3 \\
0 & 2 & 1 \\
2 & -4 & 2 \\
-3 & a & -5
\end{bmatrix}
$

Чтобы найти условие линейной зависимости, рассмотрим определитель подматрицы из первых трех строк и столбцов:

$
\det \begin{bmatrix}
1 & 2 & 3 \\
0 & 2 & 1 \\
2 & -4 & 2
\end{bmatrix} = 1(2 \cdot 2 - 1 \cdot (-4)) - 2(0 \cdot 2 - 1 \cdot 2) + 3(0 \cdot (-4) - 2 \cdot 2)  $
$
= 1(4 + 4) - 2(0 - 2) + 3(0 - 4) = 8 + 4 - 12 = 0$


Поскольку определитель равен нулю, это значит, что векторы $\mathbf{a}_1, \mathbf{a}_2, \mathbf{a}_3$ уже линейно зависимы независимо от значения $a$

Следовательно, система линейно зависима для любого значения параметра $a$. Дополнительное выражение не требуется, так как уже три вектора линейно зависимы

## Задание 5
Найдите собственные векторы и собственные значения матрицы. Найдите решение без использования Python (вручную).

$\left(\begin{array}{rrr}4 & 0 & -1 \\ 2 & 2 & 1 \\ -1 & 0 & 4\end{array}\right)$

Для нахождения собственных значений и собственных векторов матрицы сначала решим характеристическое уравнение $\det(A - \lambda I) = 0$.

1. Построение характеристического многочлена
Вычтем $\lambda I$ из $A$:

$A - \lambda I = \begin{pmatrix} 4-\lambda & 0 & -1 \\ 2 & 2-\lambda & 1 \\ -1 & 1 & 4-\lambda \end{pmatrix}$.

Найдем определитель матрицы:

$\det(A - \lambda I) = \begin{vmatrix} 4-\lambda & 0 & -1 \\ 2 & 2-\lambda & 1 \\ -1 & 1 & 4-\lambda \end{vmatrix}$.

Выполнив вычисления, получаем характеристический многочлен:

$\det(A - \lambda I) = -\lambda^3 + 10\lambda^2 - 31\lambda + 30$.

Разложим многочлен на множители:

$-\lambda^3 + 10\lambda^2 - 31\lambda + 30 = -(\lambda - 2)(\lambda^2 - 8\lambda + 15) = -(\lambda - 2)(\lambda - 3)(\lambda - 5)$.

Решив уравнение $-(\lambda - 2)(\lambda - 3)(\lambda - 5) = 0$, получаем собственные значения:

$\lambda_1 = 2$, $\lambda_2 = 3$, $\lambda_3 = 5$.

2. Поиск собственных векторов
Для $\lambda = 2$:
Подставим $\lambda = 2$ в матрицу $A - 2I$:

$A - 2I = \begin{pmatrix} 2 & 0 & -1 \\ 2 & 0 & 1 \\ -1 & 1 & 2 \end{pmatrix}$.

Решим систему уравнений $(A - 2I)v = 0$:

$\begin{cases} 2x - z = 0 \\ 2x + z = 0 \\ -x + y + 2z = 0 \end{cases}$

Из первого уравнения: $z = 2x$. Подставляем во второе: $2x + 2x = 0 \Rightarrow x = 0$. Следовательно, $z = 0$, а $y$ — свободный параметр. Получаем собственный вектор:

$v_1 = \begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}$.

Для $\lambda = 3$:
Подставим $\lambda = 3$ в $A - 3I$:

$A - 3I = \begin{pmatrix} 1 & 0 & -1 \\ 2 & -1 & 1 \\ -1 & 1 & 1 \end{pmatrix}$.

Решим систему уравнений $(A - 3I)v = 0$:

$\begin{cases} x - z = 0 \\ 2x - y + z = 0 \\ -x + y + z = 0 \end{cases}$

Из первого и третьего уравнения: $x = z$. Подставляем во второе: $2z - y + z = 0 \Rightarrow y = 3z$. Получаем собственный вектор:

$v_2 = \begin{pmatrix} 1 \\ 3 \\ 1 \end{pmatrix}$.

Для $\lambda = 5$:
Подставим $\lambda = 5$ в $A - 5I$:

$A - 5I = \begin{pmatrix} -1 & 0 & -1 \\ 2 & -3 & 1 \\ -1 & 1 & -1 \end{pmatrix}$.

Решим систему уравнений $(A - 5I)v = 0$:

$\begin{cases} -x - z = 0 \\ 2x - 3y + z = 0 \\ -x + y - z = 0 \end{cases}$

Из первого уравнения: $x = -z$. Подставляем во второе: $2(-z) - 3y + z = 0 \Rightarrow -z - 3y = 0 \Rightarrow y = -\frac{1}{3}z$. Получаем собственный вектор:

$v_3 = \begin{pmatrix} -1 \\ -\frac{1}{3} \\ 1 \end{pmatrix}$.

Итоговые результаты
Собственные значения и соответствующие собственные векторы:

$\lambda_1 = 2$, $v_1 = \begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}$,

## Задание 6
Найдите сумму сингулярных чисел для матрицы:

$A=\left(\begin{array}{ccc}-2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1\end{array}\right)$

Найдите решение без использования Python (вручную).

Найдём сумму сингулярных чисел матрицы $A$. Сингулярные числа матрицы $A$ определяются как квадратные корни собственных значений матрицы $A^TA$ (или $AA^T$).

1. Вычислим $A^TA$. Матрицы $A$ и $A^T$:

$A = \begin{bmatrix} -2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1 \end{bmatrix}$, $A^T = \begin{bmatrix} -2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1 \end{bmatrix}$.

Умножим $A^T$ на $A$:

$A^TA = \begin{bmatrix} -2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1 \end{bmatrix} \begin{bmatrix} -2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1 \end{bmatrix}$.

Вычисляем элементы новой матрицы.

$A^TA = \begin{bmatrix} 5 & 1 & -3 \\ 1 & 5 & -3 \\ -3 & -3 & 3 \end{bmatrix}$.

2. Найдём собственные значения $A^TA$. Решаем характеристическое уравнение:

$\det(A^TA - \lambda I) = 0$.

Матрица $A^TA - \lambda I$:

$A^TA - \lambda I = \begin{bmatrix} 5-\lambda & 1 & -3 \\ 1 & 5-\lambda & -3 \\ -3 & -3 & 3-\lambda \end{bmatrix}$.

Находим определитель:

$\det(A^TA - \lambda I) = -\lambda^3 + 13\lambda^2 - 36\lambda$.

Разложим на множители:

$-\lambda^3 + 13\lambda^2 - 36\lambda = -\lambda(\lambda^2 - 13\lambda + 36) = -\lambda(\lambda - 4)(\lambda - 9)$.

Собственные значения:

$\lambda_1 = 0$, $\lambda_2 = 4$, $\lambda_3 = 9$.

3. Найдём сингулярные числа.
Сингулярные числа равны квадратным корням собственных значений:

$\sigma_1 = \sqrt{\lambda_1} = 0$, $\sigma_2 = \sqrt{\lambda_2} = 2$, $\sigma_3 = \sqrt{\lambda_3} = 3$.

Сумма сингулярных чисел:

$\sigma_1 + \sigma_2 + \sigma_3 = 0 + 2 + 3 = 5$.

Ответ:
Сумма сингулярных чисел матрицы $A$ равна $5$.

## Задание 10

```python
X = data.drop(columns=["cnt"])  # все переменные, кроме целевой переменной
y = data["cnt"]  # целевая переменная

# столбец единичных значений для смещения (bias)
X = np.c_[np.ones(X.shape[0]), X]

X = np.array(X)
y = np.array(y)

# находим коэффициенты линейной регрессии
beta = np.linalg.inv(X.T @ X) @ X.T @ y

# получаем прогнозы
y_pr = X @ beta

# Среднеквадратичная ошибка
mse = np.mean((y - y_pr) ** 2)

# Коэффициент детерминации R^2
ss_total = np.sum((y - np.mean(y)) ** 2)
ss_residual = np.sum((y - y_pr) ** 2)
r2 = 1 - (ss_residual / ss_total)

print(f"MSE: {mse}")
print(f"R^2: {r2}")
```
